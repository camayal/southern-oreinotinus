### Extract all info from tiffs

```python
import glob

list_tifs = glob.glob("/home/cm2828/palmer_scratch/CHELSA_bioclim_layers/*.tif")
```

```python
list_tifs[0]
```

```python
import rasterio
rasterio.__version__
```

```python
import rasterio

def extract_chelsa_info(lat=0, long=0, list_tifs=None):
    result = {}

    result = {}
    
    for bioclim_file in list_tifs:
        bioclim_layer_name = "_".join(bioclim_file.split("/")[-1].split("_")[0:2])
        latitude = float(lat)
        longitude = float(long)
        
        with rasterio.open(bioclim_file) as dataset:
            bioclim_value = list(dataset.sample([(longitude, latitude)]))
            bioclim_number = float(bioclim_value[0][0])

        result[bioclim_layer_name] = bioclim_number
            
    return result

# dic_result = {}
# extract_chelsa_info(-16.261, -67.78667, list_tifs)
```

```python
#get database
import dbgdrive
fulldata = dbgdrive.get_database(sheet_name='sample-data', id_spreadsheet='1mnbZVtnI4AQDseWaawV2au4bNyFD_B9M0z8REMXbOcs')
sdata = fulldata[["NameInAssembly", "SearchName_Extracted", "Lat", "Long"]]

tt_result = {}
for i, r in sdata.iterrows():
    if r["NameInAssembly"] in tree.get_tip_labels():
        # print(r["NameInAssembly"], type(r["Lat"]), r["Long"])
        tt_result[r["NameInAssembly"]] = extract_chelsa_info(lat=r["Lat"], 
                                                            long=r["Long"], 
                                                            list_tifs=list_tifs)
```

```python
env_df_match_tree = pd.DataFrame(tt_result).T
```

```python
env_df_match_tree.to_csv("env_data_matching_with_tree.csv")
```

```python
# define groups using the main clades (c1 to c7)

idxs_clades = [94,89,84,79,75,68,67]
# rtree.draw(use_edge_lengths=False, node_labels=True, node_sizes=20)

cladesdict = {}
for index, node in enumerate(idxs_clades):
    tips = tree.get_tip_labels(node)
    
    # specimens = []
    for tip in tips:
        info = tip.split("_")
        specimen = info[-2] +"_"+ info[-1] 
        # specimens.append(specimen)
    
        cladesdict[tip] = f"c{index+1}"

# imap["outgroup"] = outgroup    

cladesdict
```
