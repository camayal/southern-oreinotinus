## Create IMAP

```python
#Import tree fulldataset tree with real ayava
import toytree
treeFile = f"../phylogeny/analysis-raxml/RAxML_bipartitions.10-bolivia-initial_mcov0.25_rcov0.1_ALLscaff_SelectiveSampling"
TREE = toytree.tree(treeFile)

outgroup  = ["triphyllum_Edwards_2014_04",
             "triphyllum_PWS_4011",
             "ayavacense_PWS_4002",
             "triphyllum_PWS_1769"
]

rtre = TREE.root(outgroup) #.ladderize()
```

```python
# len(incaseedum)
```

```python
## set N values of K to test across
kvalues = list(range(2,7))

## create a structure analysis object 
struct = ipa.structure(
    name=f"{NAME}",
    data=SNPS,
    mincov=0.9, 
    imap=imap,
    minmap={i: 0 for i in imap},
    # load_only=True
)

## set params for analysis
struct.mainparams.burnin = 500000
struct.mainparams.numreps = 750000
```

```python
struct.run(nreps=20, kpop=kvalues, auto=True)
```

```python
etable = struct.get_evanno_table(kvalues, quiet=False)
etable
```

```python
# get canvas object and set size
canvas = toyplot.Canvas(width=400, height=300)

# plot the mean log probability of the models in red
axes = canvas.cartesian(ylabel="Logaritmo probabilidad promedio")
axes.plot(etable.estLnProbMean * -1, color="darkred", marker="o")
axes.y.spine.style = {"stroke": "darkred"}

# plot delta K with its own scale bar of left side and in blue === In structure tuto this is the used to select the k
axes = axes.share("x", ylabel="delta K", ymax=etable.deltaK.max() + etable.deltaK.max() * .25)
axes.plot(etable.deltaK, color="steelblue", marker="o");
axes.y.spine.style = {"stroke": "steelblue"}

# set x labels
axes.x.ticks.locator = toyplot.locator.Explicit(range(len(etable.index)), etable.index)
axes.x.label.text = "K (N poblaciones ancestrales)"
```

```python
#Create namedict with last reviewd name and hold some important information like lat long
sdata = fulldata[["NameInAssembly","UltimateName","Num_for_Publication","Country","Locality","Lat","Long"]]

namedict = {}
for i in range(sdata.shape[0]):
    # part = sdata.iloc[i, 1].split("-")
    # spnum = sdata.iloc[i, 2]
    # if spnum == None:
    #     spnum = ""
    # else:
    #     spnum = " " + spnum
    namedict[sdata.iloc[i, 0]] = {
                                  #"SpecimenID": "V. " + part[0] + spnum,
                                  #"Collector": part[1],
                                  #"Country": sdata.iloc[i, 3],
                                  #"Locality": sdata.iloc[i, 4],
                                  # "AssemblyName": sdata.iloc[i, 0],
                                  "UltimateName": sdata.iloc[i, 1],
                                  "Latitude": sdata.iloc[i, 5],
                                  "Longitude": sdata.iloc[i, 6],
                                  }
```

```python
import pandas as pd
## maybe get long and use it as order measurements

def plot_all_ks(tables=tables, namedict=namedict, sorted_by="Latitude", ascending=True, save_svg=False, default_order=False):
    for k in tables:
        
        if default_order:
            # default order
            table = tables[k].sort_values(by=list(range(k)))  # sort list by columns
        else:
            # specific order
            ## create list with the order
            #get an ordered list by latitude
            _namedf = pd.DataFrame.from_dict(namedict).T
            _sorted_namedf = _namedf.sort_values(by=sorted_by, ascending=ascending)
            _samples_in_table = [name for name in tables[k].index]
            samples_ordered = _sorted_namedf[_sorted_namedf.index.isin(_samples_in_table)].index
            table = tables[k].loc[samples_ordered] # use sorted sort list by columns

        
        # update names with ultimatenames (last correct identification)
        names = [namedict[name]["UltimateName"] for name in table.index]

        # build barplot
        canvas = toyplot.Canvas(width=1000, height=250)
        axes = canvas.cartesian(bounds=("5%", "95%", "10%", "60%"))
        axes.bars(table)

        axes.x.ticks.locator = toyplot.locator.Explicit(labels=names)
        axes.x.ticks.labels.angle = -90
        axes.x.ticks.show = True
        axes.x.ticks.labels.offset = 10
        axes.x.ticks.labels.style = {"font-size": "9px"}

        best= ""
        if k == 3: best=": BEST CHOICE based on delta K"
        canvas.text(10,10, f"K{k} {best}", style={"text-anchor":"start"})

        if save_svg: toyplot.svg.render(canvas, f"./svgs/{NAME}/{NAME}_k{k}.svg")
```

```python
plot_all_ks()
```

```python
plot_all_ks(save_svg=True)
```

## Ploting one K next to the tree

```python
import toytree

# pick k to show
K = 3

# generate a random tree and data
rtre = toytree.tree("../phylogeny/analysis-raxml/RAxML_bipartitions.bolivia-initial_mcov0.25_rcov0.1_ALLscaff_ALLsamplesFound")

## because this structure was done on a subset of samples (not all samples in the tree), I am prunning the tree to be concordant with this data
table = tables[K]
rtre = rtre.prune(names=list(table.index)).root(wildcard="incarum")


# set up a toyplot Canvas with 2 axes: (x1, x2, y1, y2)
canvas = toyplot.Canvas(width=750, height=1000)
ax0 = canvas.cartesian(bounds=("5%", "80%", "5%", "100%"), padding=15) #xmin, xmax, ymin, ymax
ax1 = canvas.cartesian(bounds=("81%", "100%", "5%", "100%"), padding=15) #xmin, xmax, ymin, ymax


# get table of structure and order the structure columns by the order we have in the tree
samples_ordered = rtre.get_tip_labels()
# samples_ordered.reverse()
samples_ordered
table = tables[K].loc[samples_ordered] #replace table for one ordered with the tips in tree order


labels_updated = [namedict[i]["UltimateName"] for i in rtre.get_tip_labels()]

# add tree to first axes
rtre.draw(axes=ax0, 
          use_edge_lengths=True,
          tip_labels_align=True,
          tip_labels=labels_updated,
          node_labels="support",
    
          node_labels_style= {
            "-toyplot-anchor-shift": "10px",
            "baseline-shift": "0px",
            "text-shadow": "0.5px 0.5px #fff, -0.5px 0.5px #fff, 0.5px -0.5px #fff, -0.5px -0.5px #fff",
            "fill": "#000",
            "font-size": 8,
          },

         );
ax0.show = False

# plot the barplot on the second axes
ax1.bars(table, along='y')

# style axes
ax1.show = True
ax1.y.show = False
ax1.x.ticks.show = True

# add title
canvas.text(30,30, f"K{K}", style={"text-anchor":"start", "font-size":"30px"});
```

```python
toyplot.svg.render(canvas, f"./svgs/{NAME}/{NAME}_tree_plus_k{K}.svg")
```
