## Funtion definitions

```python
#Function that creates different imaps base on pair of species
import random

def create_imap_for_test(imap, p1, p2, p3, p4):
    """
    Given the species (keys in the imap) name, fill a dictionary with all samples 
    associated with that species (or key) in the main imap
    """
    sub_imap = {}
    sub_imap["p4"] = imap[p4]
    sub_imap["p3"] = imap[p3]
    sub_imap["p2"] = imap[p2]
#     if not p1:
#         p1 = random.choice(list(imap.keys() - [p2, p3]))       
    sub_imap["p1"] = imap[p1]
    if len(set([p1, p2, p3])) < 3: 
        raise ValueError(f"Species must be different. sp1:{p1}, sp3:{p2}, and sp3:{p3}")
    return sub_imap
```

```python
def plot_bootstraps(baba, ymin=-0.8, ymax=0.8, alpha=5.0, imap=None):
    """Draw a scatterplot of bootstrap replicate D-statistics for each test.
    
    Parameters
    ----------
    baba: ipa.baba.Baba
        An ipa.baba.Baba class instance returned by ipa.baba function and 
        that has called .run() to fill its results_table attribute.
    ymin: float
        Min domain of y-axis.
    ymax: float
        Max domain of y-axis.
    alpha: float
        Significance cutoff for Z-score to color points.
    
    Returns
    -------
    (Canvas, axes, None)
        Toyplot canvas, axes, and mark objects.
    """
    canvas = toyplot.Canvas(width=200 + 60 * baba.bootstraps.shape[0], height=300)
    axes = canvas.cartesian(xshow=False)
    axes.y.ticks.show = True

    marks = []
    for idx in range(baba.bootstraps.shape[0]):
        
        # plot box
        axes.fill(
            [idx + 0.05, idx + 0.95],
            [ymin, ymin],
            [ymax, ymax],
            style={"stroke": "black", "fill": "white"},
        )
        # plot scatter points
        fill = "black" if baba.results_table.loc[idx, "Z"] > alpha else "#D3D3D3"
        axes.scatterplot(
            np.linspace(idx + 0.1, idx + 0.9, baba.bootstraps.shape[1]),
            baba.bootstraps[idx],
            opacity=0.2,
            color=fill,
        )
    
        

        # add names to each column.

        tests = range(baba.taxon_table.shape[0])

        #define item looker
        def get_key(val):
            for key, value in imap.items():
                 if val in value:
                        return key

        #empty list for names
        names = []

        #get p1 to p3 to fill names
        for i in tests:
            p3 = baba.taxon_table.p3[i].split(",")
            p2 = baba.taxon_table.p2[i].split(",")
            p1 = baba.taxon_table.p1[i].split(",")
            #fill names
            names.append([f"({get_key(p1[0])} {get_key(p2[0])})({get_key(p3[0])})"])

        # write names
        axes.text((idx + 0.5), ymin + (ymin * 0.15), names[idx], style={"fill": "black"})
        

    # add line in zero
    axes.hlines(0, style={"stroke-dasharray": "2,4"})
    
    
    return canvas, axes, None
```

## Define main data

```python
#Create baba object with dataset of few specimens per species (with assembly names)
SNPS = "/home/cm2828/project/viburnumThings/data/Aug2022_lookingforBolivianSamples/bolivia_history_outfiles/bolivia_history.snps.hdf5"
baba_obj = ipa.baba(SNPS)
```

```python
#Import tree fulldataset tree with real ayava
treeFile = f"../phylogeny/analysis-raxml/RAxML_bipartitions.10-bolivia-initial_mcov0.25_rcov0.1_ALLscaff_SelectiveSampling"
TREE = toytree.tree(treeFile)

outgroup  = ["triphyllum_Edwards_2014_04",
             "triphyllum_PWS_4011",
             "ayavacense_PWS_4002",
             "triphyllum_PWS_1769"
]

outgroup1  = ["triphyllum_Edwards_2014_04",
             "triphyllum_PWS_4011",
]

outgroup2  = ["ayavacense_PWS_4002",
             "triphyllum_PWS_1769"
]

rtree = TREE.root(outgroup) #.ladderize()
```

```python
# define groups using the main clades (c1 to c7)

idxs_clades = [94,89,84,79,75,68,67]
# rtree.draw(use_edge_lengths=False, node_labels=True, node_sizes=20)

imap = {}
for index, node in enumerate(idxs_clades):
    imap[f"c{index+1}"] = rtree.get_tip_labels(node)

# imap["o"] = outgroup    
imap["o1"] = outgroup1   
imap["o2"] = outgroup2   

imap
```

## Ladder 1, testing p2 (nested )and p3 (nester) against outgroups split in two subgroups

If ther is some common alleles, the excess of ABBA is expected

```python
# expected excess of abba if share ancestry is stronger than derivative

imaps=[
    # Testing if there is any sharing betwen nesting and nested, z-score is interesting but maybe not as high as c6 c7
    create_imap_for_test(imap, p1= "o2", p2="c2", p3="c1", p4="o1"), 
    create_imap_for_test(imap, p1= "o2", p2="c1", p3="c2", p4="o1"), 
    # same as previous but with 2 clades that have shown similarities, here we see a big peak in z-score
    create_imap_for_test(imap, p1= "o2", p2="c7", p3="c6", p4="o1"), 
    create_imap_for_test(imap, p1= "o2", p2="c6", p3="c7", p4="o1"), 
    # testing with the rest of nesting nested clades
    create_imap_for_test(imap, p1= "o2", p2="c3", p3="c2", p4="o1"), 
    create_imap_for_test(imap, p1= "o2", p2="c2", p3="c3", p4="o1"), 
    create_imap_for_test(imap, p1= "o2", p2="c4", p3="c3", p4="o1"), 
    create_imap_for_test(imap, p1= "o2", p2="c3", p3="c4", p4="o1"), 
    create_imap_for_test(imap, p1= "o2", p2="c5", p3="c4", p4="o1"),
    create_imap_for_test(imap, p1= "o2", p2="c4", p3="c5", p4="o1"),
    create_imap_for_test(imap, p1= "o2", p2="c6", p3="c5", p4="o1"),
    create_imap_for_test(imap, p1= "o2", p2="c5", p3="c6", p4="o1"),
    # negative tests should be things that are far separated, for example c1 and c7
    create_imap_for_test(imap, p1= "o2", p2="c7", p3="c1", p4="o1"), 
    create_imap_for_test(imap, p1= "o2", p2="c1", p3="c7", p4="o1"), 


]
```

```python
minmaps=0.9
test_name = f"ladder_minmap{minmaps}"

baba_obj.run(
    imaps=imaps,
    minmaps=minmaps,
    nboots=100,
    cores=20,
    random_seed=42,
)
```

I am doing a combinatory approach to get all possible tests to see if there is some pattern. I am using here only one outgroup
